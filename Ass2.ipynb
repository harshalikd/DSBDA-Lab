```python
import pandas as pd
```

```python
data = {
    'Student_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Math_Score': [78, 85, np.nan, 92, 88, 150, 65, 70],
    'Science_Score': [72, 80, 78, np.nan, 85, 90, 60, 200],
    'English_Score': [75, 82, 79, 88, np.nan, 85, 68, 72],
    'Study_Hours': [5, 7, 6, 8, 7, 50, 4, np.nan]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df['Math_Score'].fillna(df['Math_Score'].mean(), inplace=True)
df['Science_Score'].fillna(df['Science_Score'].mean(), inplace=True)
df['English_Score'].fillna(df['English_Score'].mean(), inplace=True)
df['Study_Hours'].fillna(df['Study_Hours'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Math_Score','Science_Score','English_Score','Study_Hours']])
plt.show()
```

```python
z_scores = np.abs(stats.zscore(df[['Math_Score','Science_Score','English_Score','Study_Hours']]))
outliers = (z_scores > 3)
outliers
```

```python
for col in ['Math_Score','Science_Score','English_Score','Study_Hours']:
    upper_limit = df[col].quantile(0.95)
    lower_limit = df[col].quantile(0.05)
    df[col] = np.where(df[col] > upper_limit, upper_limit,
                        np.where(df[col] < lower_limit, lower_limit, df[col]))
```

```python
sns.boxplot(data=df[['Math_Score','Science_Score','English_Score','Study_Hours']])
plt.show()
```

```python
df['Study_Hours'].skew()
```

```python
df['Log_Study_Hours'] = np.log1p(df['Study_Hours'])
```

```python
df['Log_Study_Hours'].skew()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(df['Study_Hours'], kde=True)
plt.title("Before Transformation")

plt.subplot(1,2,2)
sns.histplot(df['Log_Study_Hours'], kde=True)
plt.title("After Log Transformation")

plt.show()
```
